```python
for i in ['PoolQC','BsmtQual','BsmtFinType2','BsmtFinType1','BsmtExposure','BsmtCond','MiscFeature','GarageFinish',
          'GarageCond','Alley','GarageQual','Fence','FireplaceQu','GarageType','MasVnrType']:
    df2[i] = df2[i].fillna("None")
    
modes = []
for i in ['Electrical']:
    modes.append(df2[i].mode()[0])
f = 0
for i in ['Electrical']:
    # print(i)
    df2[i] = df2[i].fillna(modes[f])
    f += 1

medians = []
for i in [ 'MasVnrArea', 'LotFrontage']: 
    medians.append(df2[i].median())
f = 0
for i in [ 'MasVnrArea', 'LotFrontage']: 
    df2[i] = df2[i].fillna(medians[f])
    f += 1

df2['GarageYrBlt'] = df2['GarageYrBlt'].fillna(0)    
```

```python
for i in ['PoolQC','BsmtQual','BsmtCond','BsmtFinType1','BsmtFinType2','BsmtFinType2','BsmtFinType1','BsmtExposure',
          'BsmtCond','MiscFeature','GarageFinish','GarageCond','Alley','GarageQual','Fence','FireplaceQu','GarageType',
          'MasVnrType']:
    df1[i] = df1[i].fillna("None")


modes = []
for i in ['Electrical']:
    modes.append(df1[i].mode()[0])
f = 0
for i in ['Electrical']:
    # print(i)
    df1[i] = df1[i].fillna(modes[f])
    f += 1

medians = []
for i in [ 'MasVnrArea', 'LotFrontage']: 
    medians.append(df1[i].median())
f = 0
for i in [ 'MasVnrArea', 'LotFrontage']: 
    df1[i] = df1[i].fillna(medians[f])
    f += 1

df1['GarageYrBlt'] = df1['GarageYrBlt'].fillna(0) 
```

```python
modes = []
for i in ['MSZoning','Utilities']:
    modes.append(df1[i].mode()[0])
f = 0
for i in ['MSZoning','Utilities']:
    # print(i)
    df1[i] = df1[i].fillna(modes[f])
    f += 1

df1['Exterior1st'] = df1['Exterior1st'].fillna('Other')
df1['Exterior2nd'] = df1['Exterior2nd'].fillna('Other')
df1['SaleType'] = df1['SaleType'].fillna('Oth')
df1['KitchenQual'] = df1['KitchenQual'].fillna('TA')
df1['Functional'] = df1['Functional'].fillna('Typ')

df1['BsmtFinSF1'] = df1['BsmtFinSF1'].fillna(df1['BsmtFinSF1'].median())
df1['BsmtFinSF2'] = df1['BsmtFinSF2'].fillna(df1['BsmtFinSF2'].median())
df1['BsmtUnfSF'] = df1['BsmtUnfSF'].fillna(df1['BsmtUnfSF'].median())
df1['TotalBsmtSF'] = df1['TotalBsmtSF'].fillna(df1['TotalBsmtSF'].median())
df1['BsmtUnfSF'] = df1['BsmtUnfSF'].fillna(df1['BsmtUnfSF'].median())
df1['BsmtUnfSF'] = df1['BsmtUnfSF'].fillna(df1['BsmtUnfSF'].median())
   
df1['BsmtFullBath'] = df1['BsmtFullBath'].fillna(0)
df1['BsmtHalfBath'] = df1['BsmtHalfBath'].fillna(0)
df1['GarageCars'] = df1['GarageCars'].fillna(0)
df1['GarageArea'] = df1['GarageArea'].fillna(0)
```
